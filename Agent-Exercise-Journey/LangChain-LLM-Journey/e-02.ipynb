```python
import requests

# 生成响应
def get_completion(prompt):
    try:
        # 发送请求到本地 Ollama API
        response = requests.post(
            "http://localhost:11434/api/generate",
            json={
                "model": "llama3",  # 使用下载的模型
                "prompt": prompt,
                "stream": False  # 非流式响应
            }
        )
        response.raise_for_status()  # 检查响应状态码
        # 解析响应
        return response.json()["response"]
    except requests.exceptions.RequestException as e:
        print(f"连接失败: {e}")
        return "错误: 无法连接到本地Ollama服务，请确保Ollama已启动"
    except KeyError:
        print("错误: 响应格式不正确")
        return "错误: 响应格式不正确"
    except Exception as e:
        print(f"发生错误: {e}")
        return f"错误: {str(e)}"

# 调用函数
result = get_completion("你好这是我第一次使用本地模型")
print(result)
```
