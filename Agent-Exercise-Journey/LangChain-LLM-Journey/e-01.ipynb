```python
import requests

# 设置代理
proxies = {
    "http": "http://127.0.0.1:7899",
    "https": "http://127.0.0.1:7899"
}

# 测试代理连接
try:
    response = requests.get("https://api.openai.com", proxies=proxies, timeout=5)
    print("代理连接成功！")
except Exception as e:
    print(f"代理连接失败：{e}")
```

```python
# 导入必要的库
from openai import OpenAI
import os

# 初始化OpenAI客户端
client = OpenAI(api_key=os.getenv('OPENAI_API_KEY') or 'your-api-key')

#测试连接
def get_completion(prompt):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content
# 调用函数
result = get_completion("你好这是我第一次使用OpenAI的API")
print(result)
```
